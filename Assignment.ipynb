## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import pandas as pd
  
per1 = pd.date_range(start ='1-1-2023', 
         end ='10-02-2023', freq ='M')
  
for val in per1:
    print(val)
```
